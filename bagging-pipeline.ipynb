```python
import pandas as pd
import numpy as np
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

file_path = "/kaggle/input/iris/Iris.csv"

df = pd.read_csv(file_path)
```

```python
df
```

```python
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']

le = LabelEncoder()
y_encoded = le.fit_transform(y)

y_encoded
```

```python
x_train , x_test , y_train ,y_test = train_test_split(x ,y_encoded , random_state= 42,test_size = 0.2  )
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
parameters = {
    'bagging__n_estimators': [20,50,100,150],
    'bagging__max_samples' : [0.1,0.4,0.7,1.0],
    'bagging__bootstrap': [True , False],
    'bagging__max_features': [0.1,0.4,0.6,1.0]
    
}
pip = Pipeline([('std_scaler',StandardScaler()),('bagging',BaggingClassifier(random_state = 42))])


```

```python
grid = GridSearchCV(pip , parameters,cv=5,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python


best_bagging = grid.best_estimator_
print(grid.best_score_)

```

```python

import joblib
joblib.dump(grid.best_estimator_, 'bagging_pipeline.pkl')
print("Pipeline saved! Test Accuracy:", accuracy_score(y_test, grid.best_estimator_.predict(x_test)))
```
